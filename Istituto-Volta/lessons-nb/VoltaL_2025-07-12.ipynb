## exercise 1
somma(lista) che restituisce la somma di tutti gli interi in lista usando for (non usate sum)

```python
# preparation
import random
l = [x for x in range(20)]
print(l)

# versione 1

sums=0
for el in l:
  sums+=el
print(f"usando ciclo for: {sums}")

# versione 2

print(f"usando sum(): {sum(l)}")

# versione 3

from functools import reduce
print(f"usando reduce: {reduce(lambda x,y: x+y, l)}")
```

## exercise 2
fare una funzione flatten ricorsiva per appiattire una lista annidata

```python
my_list=list()
def flatten(lista):
  for element in lista:
    if type(element)==list:
      flatten(element)
    else:
      my_list.append(element)

inserted_list = [1, 2 , [3, 4], [[9,7,8,[[23]]]]]
my_list=list()
flatten(inserted_list)
print(my_list)
```

# excercise 3
ricerca per un valore in una lista ordinata tagliando la lista a meta con la funzione ricorsiva

```python

def ricerca(lista, valore):
  if len(lista)<=1:
    return 1
  inf_lista = lista[:(len(lista)-1) //2]
  sup_lista = lista[len(inf_lista):]
  if valore in inf_lista:
    return ricerca(inf_lista, valore)
  elif valore in sup_lista:
    return len(inf_lista) + ricerca(sup_lista, valore)
  else:
    return "il valore non c'è nella lista"
print(ricerca([1,2.3,5,8,9,10,15,20,22], 5))
```
