Design and implement a Car class with the following properties:

A car has a given fuel efficiency (measured in kilometers per liter) and a certain amount of fuel in the tank.

The fuel efficiency is specified by the constructor, and the initial fuel level in the tank is zero.

Provide the following methods:

A drive method to simulate the car traveling a certain distance, reducing the fuel level accordingly;

A getGas method to check the current fuel level;

An addGas method to refuel the car.

Also, write a TestCar program that simulates a car trip.

```python
class Car:
    def _init_(self, resa):
        self.resa = resa
        self.carburante = 0

    def addGas(self, litri):
        self.carburante += litri

    def drive(self, km):
        consumo = km / self.resa
        if consumo <= self.carburante:
            self.carburante -= consumo
            print(f"Hai percorso {km} km. Carburante consumato: {consumo:.2f} litri.")
        else:
            print("Carburante insufficiente per completare il viaggio.")

    def getGas(self):
        return self.carburante


def giocoAuto():
    print("Benvenuto nel simulatore di viaggio!")
    resa = float(input("Inserisci la resa dell'auto (km per litro): "))
    auto = Car(resa)

    while True:
        print("\nScegli un'azione:")
        print("1 - Fai rifornimento")
        print("2 - Guida")
        print("3 - Controlla il carburante")
        print("4 - Esci")

        scelta = input("Scelta: ")

        if scelta == "1":
            litri = float(input("Quanti litri vuoi aggiungere? "))
            auto.addGas(litri)
            print(f"Hai aggiunto {litri} litri nel serbatoio.")

        elif scelta == "2":
            km = float(input("Quanti km vuoi percorrere? "))
            auto.drive(km)

        elif scelta == "3":
            print(f"Carburante disponibile: {auto.getGas():.2f} litri")

        elif scelta == "4":
            print("Grazie per aver giocato. A presto!")
            break

        else:
            print("Scelta non valida. Riprova.")


giocoAuto()
```
