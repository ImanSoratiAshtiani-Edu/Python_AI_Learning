# Set

creare due set chiedendo gli elementi dall'utente

```python
parole1=input("inserisci un set di parole saparate con spazi: ").split()
parole2=input("inserisci un set di parole saparate con spazi: ").split()
set_1= set(parole1)
set_2= set(parole2)
```

```python
print("set_1:" ,set_1)
print("set_2:", set_2)
```

operazioni tra i set:

```python
#difference:
print("differenza (set_1-set_2):", set_1.difference(set_2))
```

```python
# union:
print("unione degli set:", set_1.union(set_2))
```

```python
# Intersection:
print("Intersection:", set_1.intersection(set_2))
```

```python
# symmetric difference:
print("differenza_s", set_1.symmetric_difference(set_2))
```

Scrivi un programma in Python che:

Imposti una password predefinita all'interno del codice (es. "python123")

Consenta all'utente di inserire una password da tastiera

Il programma deve concedere al massimo 3 tentativi

Dopo ogni tentativo errato, il programma deve:

avvisare l’utente che la password è sbagliata

mostrare quanti tentativi rimangono

Se l’utente inserisce la password corretta:

il programma mostra un messaggio di accesso riuscito

interrompe subito il ciclo

Se l’utente esaurisce i tentativi senza successo:

il programma mostra un messaggio di accesso negato

```python
# soluzione 1
psw="Python123"
turn=0
while turn < 3:
  u_psw=input("inserisci la psw: ")
  turn+=1
  if psw == u_psw:
    print("bravo")
    quit()
  else:
    print(f"la password è errato!. hai {3-turn} tentativi da fare.")
print("hai tentato 3 volte con la psw sbagliata!")
# soluzione 2
# Password corretta predefinita
password_corretta = "python123"
tentativi = 3

# Finché ci sono tentativi
while tentativi > 0:
    tentativo = input("Inserisci la password: ")
    if tentativo == password_corretta:
        print(" Accesso riuscito!")
        break
    else:
        tentativi -= 1
        print(f" Password errata. Tentativi rimasti: {tentativi}")
else:
    print(" Accesso negato. Troppi tentativi.")
```
