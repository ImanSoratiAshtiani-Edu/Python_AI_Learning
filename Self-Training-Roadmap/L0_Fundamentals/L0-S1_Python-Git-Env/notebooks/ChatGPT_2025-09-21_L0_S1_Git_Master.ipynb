📘 Symbolic Links (لینک نمادین) چیه؟

🎯 تعریف ساده

یه میان‌بر خیلی پیشرفته هست.

درست مثل Shortcut توی ویندوزه، ولی در سطح فایل‌سیستم عمل می‌کنه (نه فقط توی Desktop).

یعنی یه فایل یا فولدر جعلی می‌سازی که وقتی روش کار می‌کنی، در واقع داری روی فایل یا فولدر اصلی کار می‌کنی.



🔹 دو نوع اصلی

Hard Link (لینک سخت):

یه اشاره‌گر مستقیم به دادهٔ واقعی روی دیسک.

فایل اصلی و hard link دقیقاً یکی هستن؛ اگه یکی رو تغییر بدی، اون یکی هم تغییر می‌کنه.

ولی اگه فایل اصلی پاک بشه، hard link هنوز داده رو نگه می‌داره (چون خودش به همون data block روی دیسک وصله).

Symbolic Link (Soft Link):

یه فایل جدید ساخته میشه که فقط آدرس فایل اصلی رو نگه می‌داره.

مثل یه تابلوی راهنما 🚦 میگه: «فایل اصلی اونجاست».

اگه فایل اصلی پاک بشه، symlink بی‌مصرف میشه (dead link).



🔹 در ویندوز

از دستور mklink استفاده میشه.
مثال:

mklink my_link.txt original.txt

حالا وقتی my_link.txt رو باز کنی، در واقع داری original.txt رو باز می‌کنی.



🔹 در لینوکس / مک

از دستور ln استفاده میشه:

ln -s original.txt my_link.txt

(-s یعنی symbolic link)



🔹 چرا استفاده می‌کنیم؟

برای صرفه‌جویی در فضا (نیازی نیست یه فایل چند بار کپی بشه).

برای سازماندهی پروژه‌ها (مثلاً یه پوشه ثابت به دیتاست، ولی می‌خوای از جای دیگه بهش دسترسی داشته باشی).

در برنامه‌نویسی و DevOps خیلی مهمه؛ مثلاً:

مدیریت نسخهٔ کتابخانه‌ها

مسیرهای اشتراکی بین پروژه‌ها

ریدایرکت کردن فایل‌های config

📘 آموزش Git با دید سیستمی و مهندسی نرم‌افزار



🔹 1. Git چیست واقعاً؟

Git یه سیستم کنترل نسخه توزیع‌شده (DVCS: Distributed Version Control System) هست.

برخلاف SVN یا CVS قدیمی که مرکزگرا بودن، Git به هر کلاینت (هر ریپوزیتوری محلی) یک کپی کامل از تاریخچه میده.

این یعنی حتی اگر سرور (GitHub/GitLab) نابود بشه، هر کلون مستقل می‌تونه دوباره به عنوان منبع (origin) عمل کنه.

فناوری‌های کلیدی:

SHA-1/SHA-256 hash: هر commit، blob، tree، و tag در Git با یک هش رمزنگاری امن شناسایی میشه. یعنی immutable (غیرقابل تغییر).

Directed Acyclic Graph (DAG): تاریخچه کامیت‌ها به شکل گراف بدون حلقه ذخیره میشه. این باعث میشه merge و branching طبیعی باشن.

Content-addressable storage: محتوای فایل‌ها و snapshotها با hash آدرس‌دهی میشن، نه اسم یا مسیر.



🔹 2. ساختار داده‌های داخلی Git

وقتی دستور git init رو می‌زنی، پشت صحنه یک دایرکتوری مخفی .git ساخته میشه.

توش اینا هست:

.git/objects/ → تمام blobها (محتوای فایل‌ها)، treeها (ساختار دایرکتوری‌ها)، commitها.

.git/refs/ → اشاره‌گرها به شاخه‌ها (branches).

.git/HEAD → میگه الان کجای تاریخچه‌ای (کدوم branch).

.git/config → تنظیمات لوکال این ریپو.



🔹 3. دستورات اصلی Git و پشت صحنه‌شون

🟢 git init

یک ریپوی خالی می‌سازه (فقط پوشه .git).

هیچ چیزی track نشده.

Option مهم:

git init --bare → فقط برای سرور. یعنی ریپوی بدون working tree (فقط تاریخچه). اینو روی سرور برای کلون‌کردن کلاینت‌ها استفاده می‌کنیم.



🟢 git add

فایل‌ها رو از working directory می‌ذاره توی staging area (index).

Git محتوای فایل رو به blob تبدیل می‌کنه (hash می‌سازه و داخل .git/objects ذخیره می‌کنه).

index فقط یه فایل باینریه (.git/index) که نگه می‌داره چه فایل‌هایی آماده‌ی commit شدن هستن.

Option مهم:

git add -p → interactive staging (تیکه‌تیکه‌ی تغییرات).

git add -u → فقط فایل‌هایی که قبلاً track بودن.

git add . → همه چیز در دایرکتوری جاری.



🟢 git commit

snapshot استیج فعلی رو برمی‌داره و یه object جدید commit می‌سازه.

commit شامل:

hash پدر (parent commit)

tree (snapshot کل پروژه در اون لحظه)

metadata (نام نویسنده، زمان، پیام)

commit immutable هست → تغییر پیام یا محتوای commit یعنی hash عوض میشه.

Option مهم:

git commit -m "msg" → پیام کوتاه.

git commit --amend → اصلاح آخرین commit (ولی hash جدید می‌سازه).

git commit --allow-empty → ساخت commit بدون تغییر (برای تست).



🟢 git status

مقایسه می‌کنه بین working directory، index و HEAD commit.

رنگ‌ها:

قرمز: تغییر کرده ولی add نشده.

سبز: staged شده (با commit بعدی ذخیره میشه).



🟢 git log

تاریخچه commitها رو نشون میده.

commitها در DAG نگه‌داری میشن.

git log --oneline --graph --decorate --all → گراف تصویری شاخه‌ها.



🟢 git branch / git checkout / git switch

branch فقط یه pointer سبک به یک commit خاصه.

git checkout قدیمی → تغییر شاخه یا فایل.

git switch (جدیدتر) → فقط برای شاخه‌ها.

Option مهم:

git branch -d → حذف شاخه.

git branch -M main → تغییر نام master به main.



🟢 git merge و git rebase

merge → تاریخچه رو گرافی نگه می‌داره. دو شاخه رو به هم وصل می‌کنه و یه commit merge می‌سازه.

rebase → commitها رو بازپخش می‌کنه روی شاخه مقصد. تاریخچه تمیز میشه ولی hashها تغییر می‌کنن.



🟢 git remote و git push/pull/fetch

git remote add origin <url> → اضافه کردن ریموت (مثلاً GitHub).

git push origin main → فرستادن commitها به سرور.

git pull = fetch + merge.

Option مهم:

git fetch --all → گرفتن همه تغییرات بدون merge.

git push -u origin main → دفعه اول، branch لوکال رو به ریموت لینک می‌کنه.



🔹 4. تکنولوژی‌های جانبی Git

Hooks: اسکریپت‌هایی که قبل/بعد از eventهایی مثل commit, push اجرا میشن (برای lint/test اتوماتیک).

Submodules: نگهداری یک ریپوی مستقل داخل ریپوی اصلی.

Worktrees: داشتن چند working directory برای یک ریپو (مثلاً برای تست چند branch همزمان).

Cherry-pick: انتقال یک commit خاص از شاخه‌ای به شاخه دیگر.



🔹 5. چرا Git اینقدر قدرتمنده؟

Immutable history → هیچ تغییری گم نمیشه (فقط hash عوض میشه).

Lightweight branches → شاخه‌ها فقط یک اشاره‌گرن، بسیار سریع.

Distributed → هر کلون یک سرور بالقوه است.

Performance → عملیات روی لوکال، نیازی به شبکه نداره.

Flexibility → merge/rebase/worktree/submodule و غیره.

📘 Commit در Git

🔹 Commit یک آبجکت است

در Git همه‌چیز آبجکت هست:

blob → محتوای یک فایل (فقط بایت‌ها، نه اسم).

tree → مثل یک فولدر، لیستی از blobها و treeهای دیگه.

commit → یک گره (node) در گراف تاریخچه، که به یک tree اشاره می‌کنه.

tag → برچسبی روی یک commit.

Commit پشت صحنه:

یک commit آبجکتیه که شامل این‌هاست:

tree <hash-of-tree>

parent <hash-of-parent-commit>

author iman <email> 1734743894 +0200

committer iman <email> 1734743894 +0200



پیام کامیت (commit message)

tree: اسنپ‌شات کل پروژه در اون لحظه (مثل «عکس لحظه‌ای»).

parent: اشاره‌گر به commit قبلی (در merge چندتا parent می‌تونه باشه).

author: کی کد رو نوشته (اسم + ایمیل + زمان).

committer: کی اون commit رو اعمال کرده (ممکنه با author فرق کنه).

پیام commit: توضیحی انسانی.

Git همه‌ی این محتوا رو می‌گیره، هش (SHA-1/SHA-256) می‌سازه، و اون هش میشه شناسه commit.



🔹 Commit در واقع تغییرات نیست!

خیلی مهم:

commit ذخیره‌ی «differences» (diff) نیست.

commit یک اشاره به یک snapshot کامل از کل پروژه‌ست.

برای کارایی، Git فقط blobهای جدید یا تغییرکرده رو ذخیره می‌کنه، ولی منطقی commit همیشه «کل وضعیت» پروژه رو نشون میده.



🔹 Commit در گراف

Commitها با هم یک Directed Acyclic Graph (DAG) می‌سازن.

هر commit به parentش اشاره می‌کنه → بنابراین زنجیره‌ی تاریخچه تشکیل میشه.

در merge commit چند parent داره (چون دو شاخه یکی شدن).



🔹 Commit از نظر سیستم فایل

اگر بری توی .git/objects/ می‌بینی که commit فقط یه فایل باینری فشرده‌شده (zlib-compressed) هست.
برای دیدنش:

این دقیقاً محتویات همون متنیه که بالا نوشتم (tree, parent, author…).



🔹 ویژگی‌های کلیدی Commit

Immutable → چون hash بر اساس کل محتواست، تغییر کوچک (حتی یک فاصله در پیام) باعث تغییر hash میشه.

Verifiable → اگر کسی تاریخچه رو دستکاری کنه، hash نمی‌خونه.

Traceable → هر commit دقیقاً یک snapshotه که میشه بازتولیدش کرد.

📘 Git Object Model + Staging Area

🔹 Git چهار نوع object داره:

blob → محتوای فایل (فقط بایت‌ها، بدون اسم).

هر فایل در پروژه تبدیل میشه به یک blob.

هش (SHA-1) از محتوا تولید میشه → شناسه blob.

اگر فایل تغییر نکنه، همون blob قبلی reuse میشه.

tree → مثل فولدر (directory).

لیست از blobها و subtreeها.

شامل نام فایل + مجوزها + hash مربوط به blob یا tree.

commit → یک اسنپ‌شات.

اشاره به یک tree داره (ریشه snapshot).

شامل parentها، author/committer، پیام commit.

tag → فقط یک reference خوانا روی commit.

lightweight tag: فقط یک اشاره.

annotated tag: object مستقل با متادیتا (مانند commit).

5) تفاوت working / index / HEAD را لمس کن

الان ۳ «لایه» داریم:

Working directory: فایل‌های روی دیسک (a.txt).

Index (staging area): آن چیزی که commit بعدی از آن عکس می‌گیرد.

HEAD: آخرین commit (snapshot قبلی).

آزمایش A — فقط working را تغییر بده

PowerShell:

"Hello Git v2" | Out-File -NoNewline a.txt

Git Bash:

echo -n "Hello Git v2" > a.txt

وضعیت:

git status

# modified: a.txt (قرمز) → یعنی working تغییر کرده اما index نه

diff نسبت به آخرین commit:

git diff           # تفاوت working vs index

git diff HEAD      # تفاوت working vs HEAD

آزمایش B — تغییر را stage کن (index را هم‌سو کن)

git add a.txt

git status

# تغییر الان سبز شده → یعنی در index هم ثبت شد

diffها:

git diff           # حالا باید خالی باشد (working == index)

git diff --cached  # تفاوت index vs HEAD را نشان می‌دهد (این هنوز تغییر کرده)

آزمایش C — commit دوم

git commit -m "update a.txt to v2"

اجزای commit جدید را ببین:

git cat-file -p HEAD

# tree <NEW-TREE-HASH>

# parent <OLD-COMMIT-HASH>

# ...

# update a.txt to v2

درخت جدید:

git cat-file -p <NEW-TREE-HASH>

# 100644 blob <NEW-BLOB-HASH>    a.txt

مشاهده: برای محتوای جدید a.txt یک blob جدید ساخته شده. tree جدید به blob جدید اشاره می‌کند. commit جدید به tree جدید و به parent (commit قبلی) اشاره می‌کند.

برخلاف SVN یا CVS قدیمی که مرکزگرا بودن، Git به هر کلاینت (هر ریپوزیتوری محلی) یک کپی کامل از تاریخچه میده.

# 📘 راهنمای مدیریت Branch و Merge Conflict در Git  

این دفترچه شامل سه بخش است:  
1. دیدن فایل‌ها و تغییرات در شاخه‌ها  
2. حذف شاخه‌های محلی و ریموت بعد از Merge  
3. مدیریت تغییرات و پیدا کردن فایل اصلی در زمان Conflict  

</div>



<div dir="rtl" align="right">

## ۱. دیدن فایل‌ها و تغییرات در شاخه‌ها  

🔹 لیست فایل‌های موجود در یک شاخه (بدون checkout):  
```bash
git ls-tree -r --name-only feat/np-ex01
git ls-tree -r --name-only feat/np-ex02
```

🔹 تفاوت فایل‌ها با شاخه main:  
```bash
git diff --name-status main..feat/np-ex01
git diff --name-status main..feat/np-ex02
```

🔹 دیدن نسخه یک فایل در شاخه‌های مختلف (بدون تغییر شاخه):  
```bash
git show main:path/to/file.py
git show feat/np-ex05:path/to/file.py
```

</div>



<div dir="rtl" align="right">

## ۲. حذف شاخه‌های محلی و ریموت بعد از Merge  

🔹 دیدن شاخه‌هایی که کاملاً merge شده‌اند:  
```bash
git checkout main
git branch --merged
```

🔹 حذف شاخه‌های محلی:  
```bash
git branch -d feat/np-ex01
git branch -d feat/np-ex02
```

🔹 حذف شاخه‌ها از GitHub (ریپوی ریموت):  
```bash
git push origin --delete feat/np-ex01
git push origin --delete feat/np-ex02
```

</div>



<div dir="rtl" align="right">

## ۳. مدیریت تغییرات و پیدا کردن فایل اصلی در زمان Conflict  

🔹 وقتی conflict رخ می‌دهد، Git در فایل این نشانه‌ها را می‌گذارد:  
```text
<<<<<<< HEAD
کدی که در شاخه main وجود داشته
=======
کدی که در شاخه feature وجود دارد
>>>>>>> feat/np-ex05
```

🔹 انتخاب نسخه شاخه اصلی (main):  
```bash
git checkout --ours path/to/file.py
```

🔹 انتخاب نسخه شاخه feature:  
```bash
git checkout --theirs path/to/file.py
```

🔹 بعد از اصلاح دستی یا انتخاب یکی از نسخه‌ها:  
```bash
git add path/to/file.py
git commit
```

🔹 لیست فایل‌هایی که conflict دارند:  
```bash
git status
```

🔹 استفاده از ابزار گرافیکی برای حل conflict:  
```bash
git mergetool
```

# 🔵 بخش ۳: استفاده از سیستم آماده برای پروژه‌های بعدی

```
poetry new --src ProjectName

```

```

poetry run pytest
```

```
git add .
git commit -m "feat: add new feature"
git push -u origin feat/new-feature
```

## ۵. Pull Request → Merge به main  
- در GitHub/GitLab یک Pull Request باز کن.  
- بعد از Merge، شاخه feature را پاک کن:  
```
git branch -d feat/new-feature
git push origin --delete feat/new-feature
```

## ۶. عملیات مهم Git (چیت‌شیت)  
```
git status                # وضعیت فعلی
git log --oneline --graph # تاریخچه
git diff                  # تفاوت‌ها
git checkout branch       # سوییچ به شاخه
git merge branch          # ادغام
git pull                  # گرفتن تغییرات remote
git push                  # ارسال تغییرات
```


- ❌ pre-commit auto-fix → دوباره add و commit:  
