````python
x = 1
for i in range(10):
  x -= .1

print(x)  # انتظار: 0، واقعیت: عددی بسیار نزدیک به صفر، اما نه دقیقاً صفر!
````

## تمرین ۳: تست مقدارهای ویژه

````python
import math
a = float('inf')
b = float('nan')
print(a > 1e308)
print(b == b)
print(math.isnan(b))
````

## decimal تمرین ۴: استفاده از



````python
from decimal import Decimal
x = Decimal('.1')
y = Decimal ('.2')
z = x + y
print(z == Decimal('.3'))
````
