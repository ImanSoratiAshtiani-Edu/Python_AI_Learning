````python
from math import prod
def custom_calculator(*args,**kwargs):
  # print (kwargs, args)
  if len(args)==0: return
  # if  kwargs == {} or kwargs['operation'] == 'sum' :
  if kwargs.get('operation', 'sum') == 'sum' :
    return sum(args)
  elif kwargs['operation'] =='multiply':
    return prod(args)
  elif kwargs['operation'] == 'average':
    return sum(args)/len(args)
  elif kwargs['operation'] == 'max':
    return max(args)
  else:
    raise ValueError("Unsupported operation")
````

````python
# correct code:
from math import prod
def custom_calculator2(*args, operation='sum'):
  if len(args)==0: return
  if operation == 'sum':
    return sum(args)
  elif operation =='multiply':
    return prod(args)
  elif operation == 'average':
    return sum(args)/len(args)
  elif operation == 'max':
    return max(args)
  else:
    raise ValueError("Unsupported operation")
````

````python
def config_summary(*args, **kwargs)->dict:
  if kwargs.get('welcome_message', False): print("\n",kwargs['welcome_message'], sep="")
  if kwargs.get('debug', False): print("Debug mode ON")
  for arg in args:
    print(arg, end=kwargs.get('end', "\n"))
    if kwargs.get('end', '\n')=='\n' and  kwargs.get('line', False):
      print(kwargs.get('line',"-")*20)
  print()
  return kwargs
````

````python
# more readable version:
def config_summary(*args, **kwargs)->dict:
  # fetch kwargs
  welcome_message = kwargs.get('welcome_message', False)
  debug= kwargs.get('debug', False)
  end = kwargs.get('end', "\n")
  line = kwargs.get('line', "-")
  # verifying conditins
  if welcome_message: print("\n",kwargs['welcome_message'], sep="")
  if debug: print("Debug mode ON")
  for arg in args:
    print(arg, end = end)
    if end =='\n':
      print(line * 20)
  print()
  return kwargs
````

````python
#testing1:
config_summary(12,15,30,50, line=".")
config_summary("iman", 15,25.0, end="\t")
config_summary("iman", 15,25.0, end=":", line=".", welcome_message="it's a very nice exercise!")
config_summary(12,15,30,50, line="=", debug=True, welcome_message="it's a very cool exercise!")
````
