#  در پایتون sys.getrefcount و ob_refcnt بررسی دقیق

## 🔍 مقدمه
در پایتون، شمارنده‌ی مرجع (Reference Count)یک عدد است که نشان می‌دهد چند متغیر یا آبجکت دیگر به یک آبجکت خاص اشاره می‌کنند. این شمارنده نقش مهمی در **مدیریت حافظه** دارد.

ماژول `sys` تابعی به نام `getrefcount()` ارائه می‌دهد که تعداد ارجاع‌های یک آبجکت را نشان می‌دهد. توجه داشته باشید که **`getrefcount()` خودش هم یک ارجاع موقتی ایجاد می‌کند.**

---

## 🔁 مثال پایه

</div>

```python
import sys

x = -548932568
y = x

print(sys.getrefcount(x))  # خروجی: 3
del y
print(sys.getrefcount(x))  # خروجی: 2
```
<div dir="rtl" align="right">



### 📌 توضیح:
- `x`: یک ارجاع
- `y`: یک ارجاع
- وقتی آرگومان به `getrefcount()` می‌دهیم: یک ارجاع موقت دیگر

پس: `3 = x + y + آرگومان`

---

## ⚠️ اشتباه رایج

اگر بنویسیم:


</div>

```
print(sys.getrefcount(-548932568))
```
<div dir="rtl" align="right">



در این صورت:

- آبجکت جدیدی با مقدار `-548932568` ساخته می‌شود.
- بنابراین، شمارنده‌ی مرجع آن چیزی نیست که به `x` یا `y` مربوط باشد.
- این مقدار تقریباً همیشه `2` خواهد بود: یکی برای آرگومان و یکی برای مقدار موقت ساخته‌شده.

## 🔬 تفاوت object ها

</div>

```python
x = -548932568
print(id(x))
print(id(-548932568))  # احتمال زیاد متفاوت است

print(x is -548932568)  # False
```

## 🧠 جمع‌بندی

- همیشه از `getrefcount(x)` استفاده کن، نه `getrefcount(مقدار)`
- عددهای کوچک مثل `-5` تا `256` در حافظه cache هستند و ممکنه نتایج غیرمنتظره بدهند
- آرگومان `getrefcount` خودش refcount را +1 می‌کند

---

## 📝 تمرین‌ها

1. بررسی کن که آیا عددهای کوچکتر از 256 در حافظه cache می‌مانند یا نه.
   - با مقایسه id و استفاده از `is`
2. کدی بنویس که:
   - لیستی از 10 عدد بسازد
   - برای هر عدد، refcount آن را چاپ کند
   - سپس همه ارجاع‌ها را حذف کند و دوباره تست کند
3. بررسی کن که هنگام append کردن یک مقدار به لیست، refcount آن مقدار چه تغییری می‌کند.
