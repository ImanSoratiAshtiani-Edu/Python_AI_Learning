### ۲. timeit با Performance تست

````python
import timeit
def plus_concat():
  result = ""
  for i in range(1000):
    result += str(i)
  return result

def join_concat():
  return "".join(str(i) for i in range(1000))

counter = 0
somma = 0
for i in range(100):
  plus = timeit.timeit(plus_concat, number = 100)
  join = timeit.timeit(join_concat, number = 100)
  somma += (plus-join)
  print((plus-join))
  if plus-join <= 0:
    counter +=1
print(somma/100)
print(counter)
````

````python
import sys, timeit


def plus_concat():
  result = ""
  for i in range(1000000):
    result += str(i)
  return result

def join_concat():
  return "".join(str(i) for i in range(1000000))

print(f"getsizeof(plus): {sys.getsizeof(plus_concat())}")
print(f"getsizeof(join): {sys.getsizeof(join_concat())}")
print(f"timeit(plus): {timeit.timeit(plus_concat, number = 100)}")
print(f"timeit(join): {timeit.timeit(join_concat, number = 100)}")
````

````python
import tracemalloc

def plus_concat():
    result = ""
    for i in range(100000):
        result += str(i)
    return result

def join_concat():
    return "".join(str(i) for i in range(100000))

# تست plus_concat
tracemalloc.start()
plus_concat()
current, peak = tracemalloc.get_traced_memory()
print(f"plus_concat: current = {current / 1024:.2f} KB, peak = {peak / 1024:.2f} KB")
tracemalloc.stop()

# تست join_concat
tracemalloc.start()
join_concat()
current, peak = tracemalloc.get_traced_memory()
print(f"join_concat: current = {current / 1024:.2f} KB, peak = {peak / 1024:.2f} KB")
tracemalloc.stop()
````

### ۱. (str) آشنایی با متدهای جادویی رشته‌ای

````python
class MyStr(str):
    def __new__(cls, value):
        return super().__new__(cls, value)

    def __repr__(self):
        return f"MyStr({super().__repr__()})"

    def __str__(self):
        return super().__str__().upper()  # نمایش بزرگ‌شده

    def __len__(self):
        return super().__len__() + 1      # مثلاً همیشه ۱ بیشتر نشون بده!

    def first3(self):
        return self[:3]

# تست
s = MyStr("python")
print(s)           # خروجی: PYTHON (به خاطر __str__)
print(repr(s))     # خروجی: MyStr('python')
print(len(s))      # خروجی: 7
print(s[2:5])      # هم مثل str معمولی اندیس‌گذاری می‌پذیره
print(s.first3())  # py
````

````python
class Modified_Str(str):
  def __str__(self):
    return super().__str__()[::-1]
  def without_vowels(self):
    tab = str.maketrans("","", "aeiou")
    return self.translate(tab)

s = Modified_Str("iman")
print(s)
print(s.without_vowels())
````

## 🔹 ۲. Indexing و Slicing

````python
s = "python"
print(s[0])     # 'p'
print(s[-1])    # 'n'
print(s[1:4])   # 'yth'
print(s[::-1])  # 'nohtyp' → معکوس
````

## 🔹 ۳. تبدیل حروف

````python
s = "ImAn"
print(s.lower())      # 'iman'
print(s.upper())      # 'IMAN'
print(s.capitalize()) # 'Iman'
print(s.title())      # 'Iman'
````

## 🔹 ۱۱. نام‌گذاری حرفه‌ای

❌ `s`, `a`, `t`  
✅ `username`, `full_name`, `cleaned_input`

````python
sentences = [
    " Python3 is cool! ",
    "data,science,2025.",
    "سلام جهان",
    "mail@openai.com",
    "Score: 98/100!!"
]

def e1():
  return [ph.strip() for ph in sentences]
def e2():
  return [ph for ph in e1() if ph.isascii()]
def e3():
  tab = str.maketrans("0123456789", "*"*10)
  return [ph.translate(tab) for ph in e1()]
def e4():
  return [ph.upper() for ph in e2() if (ph.lower().startswith("data") or ph.endswith("!"))]
def e5():
  return "+".join(ph for ph in e2())
print(e1())
print(e2())
print(e3())
print(e4())
print(e5())
````
