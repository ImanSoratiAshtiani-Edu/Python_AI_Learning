## üü¶ Subclassing str ŸÖÿ±ÿ≠ŸÑŸá €∂: ŸÖÿ™ÿØŸáÿß€å ÿ¨ÿßÿØŸà€å€å Ÿà  (Ÿæ€åÿ¥ÿ±ŸÅÿ™Ÿá Ÿà ÿßÿÆÿ™€åÿßÿ±€å)

### €±. (str) ÿ¢ÿ¥ŸÜÿß€å€å ÿ®ÿß ŸÖÿ™ÿØŸáÿß€å ÿ¨ÿßÿØŸà€å€å ÿ±ÿ¥ÿ™Ÿá‚Äåÿß€å

````python
class MyStr(str):
    def __new__(cls, value):
        return super().__new__(cls, value)

    def __repr__(self):
        return f"MyStr({super().__repr__()})"

    def __str__(self):
        return super().__str__().upper()  # ŸÜŸÖÿß€åÿ¥ ÿ®ÿ≤ÿ±⁄Ø‚Äåÿ¥ÿØŸá

    def __len__(self):
        return super().__len__() + 1      # ŸÖÿ´ŸÑÿßŸã ŸáŸÖ€åÿ¥Ÿá €± ÿ®€åÿ¥ÿ™ÿ± ŸÜÿ¥ŸàŸÜ ÿ®ÿØŸá!

    def first3(self):
        return self[:3]

# ÿ™ÿ≥ÿ™
s = MyStr("python")
print(s)           # ÿÆÿ±Ÿàÿ¨€å: PYTHON (ÿ®Ÿá ÿÆÿßÿ∑ÿ± __str__)
print(repr(s))     # ÿÆÿ±Ÿàÿ¨€å: MyStr('python')
print(len(s))      # ÿÆÿ±Ÿàÿ¨€å: 7
print(s[2:5])      # ŸáŸÖ ŸÖÿ´ŸÑ str ŸÖÿπŸÖŸàŸÑ€å ÿßŸÜÿØ€åÿ≥‚Äå⁄Øÿ∞ÿßÿ±€å ŸÖ€å‚ÄåŸæÿ∞€åÿ±Ÿá
print(s.first3())  # py
````

## üîπ €±€±. ŸÜÿßŸÖ‚Äå⁄Øÿ∞ÿßÿ±€å ÿ≠ÿ±ŸÅŸá‚Äåÿß€å

‚ùå `s`, `a`, `t`  
‚úÖ `username`, `full_name`, `cleaned_input`

````python
sentences = [
    " Python3 is cool! ",
    "data,science,2025.",
    "ÿ≥ŸÑÿßŸÖ ÿ¨ŸáÿßŸÜ",
    "mail@openai.com",
    "Score: 98/100!!"
]

def e1():
  return [ph.strip() for ph in sentences]
def e2():
  return [ph for ph in e1() if ph.isascii()]
def e3():
  tab = str.maketrans("0123456789", "*"*10)
  return [ph.translate(tab) for ph in e1()]
def e4():
  return [ph.upper() for ph in e2() if (ph.lower().startswith("data") or ph.endswith("!"))]
def e5():
  return "+".join(ph for ph in e2())
print(e1())
print(e2())
print(e3())
print(e4())
print(e5())
````
