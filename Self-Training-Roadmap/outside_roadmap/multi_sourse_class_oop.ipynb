````python
class Point:
    def __init__(self,x,y): # constructor or initializer method
        self.x=x
        self.y=y
    def draw(self):         # Instance method
        print(f"Point({self.x}, {self.y})")
    @classmethod            # <--
    def zero(cls):          # factory method (class method)
        return cls(0,0)
point=Point.zero()          # factory method can create a factory object
point.draw()
print(point)
print(str(point))
````

### طراحی یک ساختمان داده شخصی

````python
class BagOfWords:
    def __init__(self):
        self.words=dict()
    def add(self,word):
        self.words[word.capitalize()]=self.words.get(word.capitalize(),0)+1
    def __getitem__(self, word):        # used with "word[]"
        return self.words.get(word.capitalize(),0)
    def __setitem__(self,word,count):   # used with "word[] = count"
        self.words[word.capitalize()]=count
    def __iter__(self):                 # used with "for statement"
        return iter(self.words)
    def __len__(self):
        return len(self.words)          # used with "len(object)""
    def __str__(self):
        from prettytable import PrettyTable as Table
        stampa=Table()
        stampa.add_column("word",list(self.words.keys()))
        stampa.add_column("count",list(self.words.values()))
        return stampa.get_string()
````

````python
class Product:
    def __init__(self, price):
        self.set_price(price)           # <--
    def get_price(self):
        return self.__price
    def set_price(self,value):
        if value < 0:
            raise ValueError("Negative price is not allowed")  # <--
        self.__price=value              # <--
````

````python
product1=Product(10)
print(product1.get_price())
product2=Product(-10)
````

ب: با پراپرتی به روش نا مناسب

````python
class Product:
    def __init__(self, price):
        self.__set_price(price)
    def __get_price(self):                      # private method
        return self.__price
    def __set_price(self,value):                # private method
        if value < 0:
            raise ValueError("Negative price is not allowed")
        self.__price=value

    price=property(__get_price,__set_price)     # is added this line
````

````python
product3=Product(10)
#print(product3.__get_price())  #AttributeError: 'Product' object has no attribute '__get_price'
print(product3.price)
product24=Product(-10)
````

  ب: با پراپرتی با دکوریتور

````python
class Product:
    def __init__(self, price):
        self.price= price           # here we behave it as a normal attribute not a private one
    @property
    def price(self):                # price is a property
        return self.__price         # inside setter and getter, we have __
    @price.setter                   # the property will be set here
    def price(self,value):
        if value < 0:
            raise ValueError("Negative price is not allowed")
        self.__price=value          # inside setter and getter, we have __
````

````python
product5=Product(10)
product5.price = 12
#print(product5.price())  #TypeError: 'int' object is not callable
print(product5.price)
product6=Product(-10)
````

````python
class MyClass:
    def __init__(self, value):
        self._value = value

    @property
    def value(self):
        return self._value

    @value.setter
    def value(self, val):
        if val >= 0:
            self._value = val
        else:
            raise ValueError("value باید مثبت باشد")

    @value.deleter
    def value(self):
        del self._value
````

````python
from abc import ABC, abstractmethod

# create a personal Error class
class InvalidOperationError(Exception):
  pass

class Stream(ABC):                    #<--
  def __init__(self):
    self.opened=False
  def open(self):
    if self.opened:
      raise InvalidOperationError("The Stream is already opened")
    self.opened=True
    print("The stream is opening")
  def close(self):
    if not self.opened:
      raise InvalidOperationError("The Stream is not opened")
    self.opened=False
    print("The stream is closing")

  @abstractmethod
  def read(self):
    pass

class FileStream(Stream):
  def read(self):
    print("The stream is reading from a file")

class UrlStream(Stream):
  def read(self):
    print("The stream is reading from a URL")
````

````python
from abc import ABC, abstractmethod
import math
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Circle(Shape):
  def __init__(self, r):
    self.r=r
  def area(self):
    print("Area of circle({}): {}".format(self.r,2*math.pi*self.r))

class Square(Shape):
  def __init__(self, x):
    self.x=x
  def area(self):
    print(f"Area of square({self.x}): {self.x**2}")
````

````python
def areaOfShape(shape):
  shape.area()  # polymorphism
````

````python
square=Square(5)
circle=Circle(5)
areaOfShape(square)
areaOfShape(circle)
````

````python
def area(shapes):
  for shape in shapes:
    shape.area()  # polymorphism
````

````python
area((square,circle))
````

آبجکت های گرفته شده قابل تغییر نخواهند بود

````python
from collections import namedtuple
Person=namedtuple("Person",("name", "age"))
p1=Person("iman",40)
p2=Person("iman",40)
````

````python
print(p1==p2)
print(id(p1), id(p2))
print(p1)
try:
  p1.name="magic"
except AttributeError as E:
  print("Error: ",E)
````

````python
from dataclasses import dataclass

@dataclass
class Person:
  name: str
  age: int
````

````python
p1=Person("iman", 40)
p2=Person("iman", 40)
print(p1==p2)
print(id(p1))
print(id(p2))

p1.name="magic"
print(p1)

try:
  print(p1>p2)
except TypeError as E:
  print("Error: ",E)
````

````python
from dataclasses import dataclass

@dataclass(frozen=True)
class Point:
  x: int
  y: int
````

````python
p1=Point(1,5)
p2=Point(3,7)
print(p2)
print(p1)
try:
  p1.x="2"
except AttributeError as E:
  print("Error: ",E)
try:
  print(p1>p2)
except TypeError as E:
  print("Error: ",E)
````

````python
from dataclasses import dataclass

@dataclass(frozen=True, order=True)
class Point:
  x: int
  y: int
````

````python
# test
p1=Point(1,50)
p2=Point(3,7)
print(p2)
print(p1)
try:
  p1.x="2"
except AttributeError as E:
  print("Error: ",E)
try:
  print(p1<p2)
except TypeError as E:
  print("Error: ",E)
````

````python
import random

class List:
    def __init__(self):
        self.values = []

    def create(self, n, lbound, ubound):
        self.n = n
        self.lbound = lbound
        self.ubound = ubound
        self.values = random.sample(range(self.lbound, self.ubound), k=self.n)

    def __str__(self):
        return str(self.values)

# Test
l1 = List()
l1.create(7, 100, 1000)
print(l1)
print(type(l1))
#l1.append(1)    # error because l1 is an obj of List and don't have append() as method
````
