### Instance method vs class methods

````python
class Point:
    def __init__(self,x,y): # constructor or initializer method
        self.x=x
        self.y=y
    def draw(self): # Instance method
        print(f"Point({self.x}, {self.y})")
````

````python
point=Point(0,0)
point.draw()
````

factory method and factory object

````python
class Point:
    def __init__(self,x,y): # constructor or initializer method
        self.x=x
        self.y=y
    def draw(self): # Instance method
        print(f"Point({self.x}, {self.y})")
    @classmethod
    def zero(cls):
        return cls(0,0)
point=Point.zero() # factory method can create a factory object
point.draw()
print(point)
print(str(point))
````

#### Representing Objects as Strings

رفتاری است که آبجکت کلاس نشان میدهد وقتی آبجکت را پرینت میگیریم

User-Friendly String Representations With .__str__()

Python calls this method when you call the built-in str() function, passing an instance of the class as an argument.

Python also calls this method when you use the instance as an argument to the print() and format() functions.

@https://realpython.com/python-magic-methods/#representing-objects-as-strings

````python
class Point:
    def __init__(self,x,y): # constructor or initializer method
        self.x=x
        self.y=y
    def draw(self): # Instance method
        print(f"Point({self.x}, {self.y})")
    @classmethod
    def zero(cls):
        return cls(0,0)
    def __str__(self):
        return f"({self.x}, {self.y})"
point=Point.zero() # factory method can create a factory object
````

````python
print(point)
print(str(point))
````

### طراحی یک ساختمان داده شخصی

````python
class BagOfWords:
    def __init__(self):
        self.words=dict()
    def add(self,word):
        self.words[word.capitalize()]=self.words.get(word.capitalize(),0)+1
    def __getitem__(self, word):
        return self.words.get(word.capitalize(),0)
    def __setitem__(self,word,count):
        self.words[word.capitalize()]=count
    def __iter__(self):
        return iter(self.words)
    def __len__(self):
        return len(self.words)
    def __str__(self):
        from prettytable import PrettyTable as Table
        stampa=Table()
        stampa.add_column("word",list(self.words.keys()))
        stampa.add_column("count",list(self.words.values()))
        return stampa.get_string()
````

````python
document=BagOfWords()
# print(document)
document.add("iman")
print(len(document))
print(document["giulia"])
document["giulia"]=3
document.add("IMAN")
print(document)
for word in document:
    print(word,document[word])
````

#### property


پراپرتی در پایتون یعنی در گنترل گرفتن مقدار دهی، خواندن و حذف اتریبیوت.
برای این کار لازم است اتریبیوت پرایوت شده باشد.

مثال: میخواهیم که اتریبیوت ما مقدار منفی نگیرد

الف: بدون پراپرتی

````python
class Product:
    def __init__(self, price):
        self.set_price(price)           # <--
    def get_price(self):
        return self.__price
    def set_price(self,value):
        if value < 0:
            raise ValueError("Negative price is not allowed")  # <--
        self.__price=value              # <--
````

````python
product1=Product(10)
print(product1.get_price())
product2=Product(-10)
````

ب: با پراپرتی به روش نا مناسب

````python
class Product:
    def __init__(self, price):
        self.__set_price(price)
    def __get_price(self):                      # private method
        return self.__price
    def __set_price(self,value):                # private method
        if value < 0:
            raise ValueError("Negative price is not allowed")
        self.__price=value

    price=property(__get_price,__set_price)     # is added this line
````

````python
product3=Product(10)
#print(product3.__get_price())  #AttributeError: 'Product' object has no attribute '__get_price'
print(product3.price)
product24=Product(-10)
````

  ب: با پراپرتی با دکوریتور

````python
class Product:
    def __init__(self, price):
        self.price= price           # here we behave it as a normal attribute not a private one
    @property
    def price(self):                # price is a property
        return self.__price         # inside setter and getter, we have __
    @price.setter                   # the property will be set here
    def price(self,value):
        if value < 0:
            raise ValueError("Negative price is not allowed")
        self.__price=value          # inside setter and getter, we have __
````

````python
product5=Product(10)
product5.price = 12
#print(product5.price())  #TypeError: 'int' object is not callable
print(product5.price)
product6=Product(-10)
````

````python
import random

class List:
    def __init__(self):
        self.values = []

    def create(self, n, lbound, ubound):
        self.n = n
        self.lbound = lbound
        self.ubound = ubound
        self.values = random.sample(range(self.lbound, self.ubound), k=self.n)

    def __str__(self):
        return str(self.values)

# Test
l1 = List()
l1.create(7, 100, 1000)
print(l1)
print(type(l1))
#l1.append(1)    # error because l1 is an obj of List and don't have append() as method
````
