### ✅ روند پیشنهادی:
<div align="left"><a href="#top">⬆️ بازگشت به بالا</a></div>

از train.csv استفاده کن و با train_test_split اون رو به دو بخش تقسیم کن.

```python
from sklearn.model_selection import train_test_split

# جدا کردن ویژگی‌ها و برچسب‌ها
X = df.drop('Survived', axis = 1)
y = df['Survived']

# تقسیم داده‌ها به آموزش و تست
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=42)
```

### الگوریتم‌های مناسب: DecisionTree, RandomForest, LogisticRegression, KNN

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

```

<div dir="rtl" align="right">

<a id="sec-022"></a>
## 🔹 ۶. آموزش مدل
<div align="left"><a href="#top">⬆️ بازگشت به بالا</a></div>
</div>


```python
# آموزش مدل رگرسیون لجستیک  
model_lr= LogisticRegression(random_state=42)
model_lr.fit(X_train, y_train)
y_lr=model_lr.predict(X_test)
model_lr
```

```python
# knn آموزش مدل 
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(X_train, y_train)
y_knn=model_knn.predict(X_test)
model_knn
```

```python
# آموزش مدل درخت تصمیم
model_dt = DecisionTreeClassifier(random_state=42, max_depth=4)
model_dt.fit(X_train, y_train)
y_dt=model_dt.predict(X_test)
model_dt
```

```python
model_rf=RandomForestClassifier(n_estimators = 100, random_state=42, max_depth=4)
model_rf.fit(X_train, y_train)
y_rf=model_rf.predict(X_test)
model_rf
```

## 🔹 ۷. ارزیابی مدل
<div align="left"><a href="#top">⬆️ بازگشت به بالا</a></div>
</div>


```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
def evaluate_model(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    class_report = classification_report(y_true, y_pred)
    conf_matrix = confusion_matrix(y_true, y_pred)
    return accuracy, class_report, conf_matrix
 
```

```python
from prettytable import PrettyTable
table=PrettyTable()
table.field_names=["Model","Accuracy","Classification Report","Confusion Matrix"]

table.add_row(("Logistic Regression", *evaluate_model(y_test, y_lr)))
table.add_row(("K-Nearest Neighbors", *evaluate_model(y_test, y_knn)))
table.add_row(("Decision Tree", *evaluate_model(y_test, y_dt)))
table.add_row(("Random Forest", *evaluate_model(y_test, y_rf)))

print(table)
```

```python
def plot_conf_matrix(y_true, y_pred, model_name):
    plt.figure(figsize=(5,4))
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.savefig(f"../images/conf_matrix_{model_name.replace(' ', '_').lower()}.png")
    plt.show()

# رسم برای هر مدل
plot_conf_matrix(y_test, y_lr, "Logistic Regression")
plot_conf_matrix(y_test, y_knn, "K-Nearest Neighbors")
plot_conf_matrix(y_test, y_dt, "Decision Tree")
plot_conf_matrix(y_test, y_rf, "Random Forest")
```

#### 📊 رسم نمودار ROC و محاسبه AUC
<div align="left"><a href="#top">⬆️ بازگشت به بالا</a></div>
</div>


<div dir="rtl" align="right">
برای این کار، باید از نسخه‌های احتمالی پیش‌بینی‌ها استفاده کنیم، نه فقط predict() بلکه predict_proba():
</div>

```python
from sklearn.metrics import roc_curve, auc

def plot_roc_curve(y_true, y_proba, model_name):
    fpr, tpr, _ = roc_curve(y_true, y_proba)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {model_name}')
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.savefig(f'../images/roc_{model_name.replace(" ", "_").lower()}.png')
    plt.show()

# گرفتن احتمال‌ها
y_lr_proba = model_lr.predict_proba(X_test)[:,1]
y_knn_proba = model_knn.predict_proba(X_test)[:,1]
y_dt_proba = model_dt.predict_proba(X_test)[:,1]
y_rf_proba = model_rf.predict_proba(X_test)[:,1]

# رسم نمودارها
plot_roc_curve(y_test, y_lr_proba, "Logistic Regression")
plot_roc_curve(y_test, y_knn_proba, "K-Nearest Neighbors")
plot_roc_curve(y_test, y_dt_proba, "Decision Tree")
plot_roc_curve(y_test, y_rf_proba, "Random Forest")
```

<div dir="rtl" align="right">

وقتی داده‌ها نامتوازن (imbalanced) هستن—یعنی تعداد نمونه‌ها در کلاس‌های مختلف تفاوت زیادی دارن—استفاده از accuracy به عنوان معیار ارزیابی (scoring) می‌تونه گمراه‌کننده باشه. چون مدل ممکنه فقط کلاس غالب رو پیش‌بینی کنه و همچنان دقت بالایی نشون بده، در حالی که عملکرد واقعی‌اش ضعیفه.

* f1 یا f1_macro:

ترکیبی از precision و recall.

f1_macro: میانگین F1 برای همه کلاس‌ها، بدون توجه به تعداد نمونه‌ها.

مناسب برای داده‌های نامتوازن.

مخصوص مدل‌های باینری یا چندکلاسه.

عملکرد مدل در تفکیک کلاس‌ها رو نشون می‌ده.



* balanced_accuracy:

میانگین دقت برای هر کلاس.

نسبت به عدم توازن حساس نیست.

</div>
